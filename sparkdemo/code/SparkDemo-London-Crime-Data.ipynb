```python
from pyspark.sql import SparkSession
import pyspark.sql.functions as func
```

```python
spark = SparkSession.builder\
                    .appName("Analysing London crime data")\
                    .getOrCreate()
```

```python
data = spark.read\
            .format("csv")\
            .option("header", "true")\
            .load("../datasets/london_crime_by_lsoa.csv")
```

```python
data.printSchema()
```

```python
data.count()
```

```python
data.limit(5).show()
```

```python
data.dropna()
```

```python
data = data.drop("lsoa_code")

data.show(5)
```

```python
total_boroughs = data.select('borough')\
                    .distinct()
total_boroughs.show()
```

```python
hackney_data = data.filter(data['borough'] == "Hackney")
hackney_data.show(5)
```

```python
data_2015_2016 = data.filter(data['year'].isin("2015", "2016"))
data_2015_2016.sample(fraction=0.1).show()
```

```python
data_greater_than_2014 = data.filter(data['year'] > "2014")
data_greater_than_2014.sample(fraction=0.1).show()
```

```python
borough_crime_count = data.groupBy('borough')\
                            .count()
borough_crime_count.show(5)
```

```python
borough_conviction_sum = data.groupBy('borough')\
                            .agg({"value":"sum"})\
                            .withColumnRenamed("sum(value)","convictions")
borough_conviction_sum.show(5)
```

```python
total_borough_convictions = borough_conviction_sum.agg({"convictions":"sum"})
total_borough_convictions.show()
```

```python
total_convictions = total_borough_convictions.collect()[0][0]
```

```python
borough_percentage_contribution = borough_conviction_sum.withColumn(
    "% contribution",
    func.round(borough_conviction_sum.convictions / total_convictions * 100, 2))
borough_percentage_contribution.printSchema()
```

```python
borough_percentage_contribution.orderBy(borough_percentage_contribution[2].desc())\
                                .show(10)
```

```python
conviction_monthly = data.filter(data['year'] == 2014)\
                        .groupBy('month')\
                        .agg({"value":"sum"})\
                        .withColumnRenamed("sum(value)","convictions")\
                        .orderBy("convictions", ascending=0)
conviction_monthly.show()
```

```python
total_conviction_monthly = conviction_monthly.agg({"convictions":"sum"})\
                                                .collect()[0][0]
total_conviction_monthly = conviction_monthly.withColumn(
                            "percent",
                            func.round(conviction_monthly.convictions / total_conviction_monthly * 100, 2))
total_conviction_monthly.columns
total_conviction_monthly.show()
```

```python
crimes_category = data.groupBy('major_category')\
                        .agg({"value":"sum"})\
                        .withColumnRenamed("sum(value)", "convictions")
crimes_category.orderBy("convictions", ascending=0).show()
```

```python
# This is how you get inline plots with matplotlib IMPORTANT
get_ipython().magic('matplotlib inline')
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
# this is a function that can be used to plot convictions by borough for a certain year
def describe_year(year):
    yearly_details = data.filter(data.year == year)\
                        .groupBy('borough')\
                        .agg({"value":"sum"})\
                        .withColumnRenamed("sum(value)", "convictions")
    
    # loop through the data frame and take the first column into an array
    borough_list = [x[0] for x in yearly_details.toLocalIterator()]
    # loop through the data frame and take the second column into an array
    convictions_list = [x[1] for x in yearly_details.toLocalIterator()]
    
    plt.figure(figsize=(33,10))
    plt.bar(borough_list, convictions_list)
    
    plt.title('Crime for the year: ' + str(year), fontsize=30)
    plt.xlabel('Boroughs', fontsize=30)
    plt.ylabel('Convictions', fontsize=30)
    
    plt.xticks(rotation=90, fontsize=30)
    plt.yticks(fontsize=30)
    plt.autoscale()
    plt.show()
```

```python
describe_year(2014)
```

<p>
    <b>Closure Definition</b> - a closure is this nested function that has been returned and assigned to a variable by a calling function
</p>

<p>
    <b>Closure Info</b> - The closure still has access to the local variables once it is returned to the calling function and assigned to a variable. The closure creates its own copies of the variables and takes it with it once it is returned.
</p>
